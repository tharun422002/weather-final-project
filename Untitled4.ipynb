```python
data = pd.read_csv('airline-passenger-traffic(1).csv', header = None)
data.columns = ['Month','Passengers']
data['Month'] = pd.to_datetime(data['Month'], format='%Y-%m')
data = data.set_index('Month')
data.shape
data
```

```python
data.plot(figsize=(20, 5))
plt.grid()
plt.legend(loc='best')
plt.title('Airline passenger traffic')
plt.show(block=False)
```

```python
data = data.assign(Passengers_Linear_Interpolation=data.Passengers.interpolate(method='linear'))
data[['Passengers_Linear_Interpolation']].plot(figsize=(20, 5))
plt.grid()
plt.legend(loc='best')
plt.title('Airline passenger traffic: Linear interpolation')
plt.show(block=False)
```

```python
data['PassengersInterpolation'] = data['Passengers_Linear_Interpolation']
data.drop(columns=['Passengers_Linear_Interpolation'],inplace=True)
```

```python
original = data.drop(['Passengers'],axis=1)
```

```python
original
```

```python
train_len = 120
train = original[0:train_len] # first 120 months as training set
test = original[train_len:] # last 24 months as out-of-time test set
```

```python
train.shape
```

```python
test.shape
```

```python
from statsmodels.tsa.stattools import adfuller #Statistics test adfuller
adf_test = adfuller(original['PassengersInterpolation'])

print('ADF Statistic: %f' % adf_test[0])
print('Critical Values @ 0.05: %.2f' % adf_test[4]['5%'])
print('p-value: %f' % adf_test[1])
```

```python
from scipy.stats import boxcox
data_boxcox = pd.Series(boxcox(original['PassengersInterpolation'], lmbda=0), index = data.index)

# data['Passengers'].plot(figsize=(20, 5))
# plt.grid()
plt.plot(data_boxcox, label='After Box Cox tranformation')
plt.legend(loc='best')
plt.title('After Box Cox transform')
plt.show()
```

```python
data_boxcox
```

```python
adf_test = adfuller(data_boxcox)

print('ADF Statistic: %f' % adf_test[0])
print('Critical Values @ 0.05: %.2f' % adf_test[4]['5%'])
print('p-value: %f' % adf_test[1])
```

```python
from statsmodels.graphics.tsaplots import plot_pacf
plt.figure(figsize=(20,5))
plt.grid()
plot_pacf(data_boxcox, ax=plt.gca(), lags = 30)
plt.show()
```

```python
from pandas import read_csv
from matplotlib import pyplot
from statsmodels.tsa.ar_model import AutoReg
from sklearn.metrics import mean_squared_error
from math import sqrt
```

```python
train_data_boxcox = data_boxcox[:train_len]
test_data_boxcox = data_boxcox[train_len:]
```

```python
model = AutoReg(train_data_boxcox, lags=29)
model_fit = model.fit()
#print('Coefficients: %s' % model_fit.params)
# make predictions
predictions = model_fit.predict(start=len(train_data_boxcox), end=len(train_data_boxcox)+len(test_data_boxcox)-1, dynamic=False)
print(predictions)
```

```python
rmse = sqrt(mean_squared_error(test_data_boxcox, predictions))
print(rmse)
```
